```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Total 'goal inside the penalty' for each team





```python
inside1=df.groupby('team1')['goal inside the penalty area team1'].sum()
inside2=df.groupby('team2')['goal inside the penalty area team2'].sum()
inside=inside1+inside2
new_1=pd.DataFrame(inside).reset_index()
new_1.rename({0: 'goal inside the penalty'},
          axis = "columns", inplace = True)
```

Total 'goal outside the penalty area' for each team


```python
outside1=df.groupby('team1')['goal outside the penalty area team1'].sum()
outside2=df.groupby(['team2'])['goal outside the penalty area team2'].sum()
outside=outside1+outside2
new_2=pd.DataFrame(outside).reset_index()
new_2.rename({0: 'goal outside the penalty area'},
          axis = "columns", inplace = True)
```

Total 'yellow cards' for each team

```python
ycards1=df.groupby('team1')['yellow cards team1'].sum()
ycards2=df.groupby('team2')['yellow cards team2'].sum()
ycards = ycards1+ycards2
new_3=pd.DataFrame(ycards).reset_index()
new_3.rename({0: 'yellow cards'},
          axis = "columns", inplace = True)
```

Total 'red cards' for each team

```python
cards_r_1=df.groupby('team1')['red cards team1'].sum()
cards_r_2=df.groupby('team2')['red cards team2'].sum()
rcards = cards_r_1+cards_r_2
new_4=pd.DataFrame(rcards).reset_index()
new_4.rename({0: 'red cards'},
          axis = "columns", inplace = True)
```

Average 'total attempts' for each team

```python
attempts1=df.groupby('team1')['total attempts team1'].mean()
attempts2=df.groupby('team2')['total attempts team2'].mean()
Avg_attempts=(attempts1+attempts2)/2
new_5=pd.DataFrame(Avg_attempts).reset_index()
new_5.rename({0: 'total attempts'},
          axis = "columns", inplace = True)
```

Average 'on target attempts' for each team

```python
attempts_on_1=df.groupby('team1')['on target attempts team1'].mean()
attempts_on_2=df.groupby('team2')['on target attempts team2'].mean()
Avg_attempts_on_target=(attempts_on_1+attempts_on_2)/2
new_6=pd.DataFrame(Avg_attempts_on_target).reset_index()
new_6.rename({0: 'on target attempts'},
          axis = "columns", inplace = True)
```

Average 'off target attempts' for each team

```python
attempts_off_1=df.groupby('team1')['off target attempts team1'].mean()
attempts_off_2=df.groupby('team2')['off target attempts team2'].mean()
Avg_attempts_off_target=(attempts_off_1+attempts_off_2)/2
new_7=pd.DataFrame(Avg_attempts_off_target).reset_index()
new_7.rename({0: 'off target attempts'},
          axis = "columns", inplace = True)
```

Average 'attempts inside the penalty area' for each team

```python
attempts_inside_1=df.groupby('team1')['attempts inside the penalty area team1'].mean()
attempts_inside_2=df.groupby('team2')['attempts inside the penalty area  team2'].mean()
Avg_attempts_inside_penalty_area=(attempts_inside_1+attempts_inside_2)/2
new_8=pd.DataFrame(Avg_attempts_inside_penalty_area).reset_index()
new_8.rename({0: 'attempts inside the penalty area'},
          axis = "columns", inplace = True)
```

Average 'attempts outside the penalty area' for each team

```python
attempts_outside_1=df.groupby('team1')['attempts outside the penalty area  team1'].mean()
attempts_outside_2=df.groupby('team2')['attempts outside the penalty area  team2'].mean()
Avg_attempts_outside_penalty_area=(attempts_outside_1+attempts_outside_2)/2
new_9=pd.DataFrame(Avg_attempts_outside_penalty_area).reset_index()
new_9.rename({0: 'attempts outside the penalty area'},
          axis = "columns", inplace = True)
```

Average 'Conceded' for each team

```python
con1 =df.groupby('team1')['conceded team1'].mean()
con2=df.groupby('team2')['conceded team2'].mean()
con=con1+con2
new_10=pd.DataFrame(con).reset_index()
new_10.rename({0 : 'Conceded'},
          axis = "columns", inplace = True)
```

Average 'Assists' for each team

```python
assist1 =df.groupby('team1')['assists team1'].mean()
assist2=df.groupby('team2')['assists team2'].mean()
assist = assist1+ assist2
new_11=pd.DataFrame(assist).reset_index()
new_11.rename({0 : 'Assits'},
          axis = "columns", inplace = True)
```

Total 'Passes completed' for each team

```python
pass1 =df.groupby('team1')['passes completed team1'].sum()
pass2=df.groupby('team2')['passes completed team2'].sum()
pass_total = pass1 + pass2
new_12=pd.DataFrame(pass_total).reset_index()
new_12.rename({0 : 'Passes completed'},
          axis = "columns", inplace = True)
```

Total 'Corners' for each team

```python
corn1 =df.groupby('team1')['corners team1'].sum()
corn2=df.groupby('team2')['corners team2'].sum()
corn = corn1+corn2
new_13=pd.DataFrame(corn).reset_index()
new_13.rename({0 : 'Corners team'},
          axis = "columns", inplace = True)
```

Average 'Free kicks' for each team

```python
free_kicks_1 = df.groupby('team1')['free kicks team1'].mean()
free_kicks_2 = df.groupby('team2')['free kicks team2'].mean()
Avg_free_kicks = (free_kicks_1 + free_kicks_2)/2
new_15 = pd.DataFrame(Avg_free_kicks).reset_index()
new_15.rename({0 : 'Free kicks'},
          axis = "columns", inplace = True)
```

Total 'Completed defensive line breaks' for each team




```python
cdlb1 =df.groupby('team1')['completed defensive line breaksteam1'].sum()
cdlb2=df.groupby('team2')['completed defensive line breaks team2'].sum()
cdlb = cdlb1 +cdlb2
new_16=pd.DataFrame(cdlb).reset_index()
new_16.rename({0 : 'Completed defensive line breaks'},
          axis = "columns", inplace = True)
```

Total 'Offsides' for each team


```python
offside1 =df.groupby('team1')['offsides team1'].sum()
offside2=df.groupby('team2')['offsides team2'].sum()
offside = offside1+offside2
new_17=pd.DataFrame(offside).reset_index()
new_17.rename({0 : 'Offsides team'},
          axis = "columns", inplace = True)
```

Total 'Attempted defensive line breaks' for each team


```python
attempted_defensive_1 = df.groupby('team1')['attempted defensive line breaks team1'].sum()
attempted_defensive_2 = df.groupby('team2')['attempted defensive line breaks team2'].sum()
total = attempted_defensive_1 + attempted_defensive_2
new_18 = pd.DataFrame(total).reset_index()
new_18.rename({0 : 'Total attempted defensive line breaks'},
          axis = "columns", inplace = True)
```

Total 'Goals' for each team


```python
part1 =df.groupby('team1')['number of goals team1'].sum()
part2=df.groupby('team2')['number of goals team2'].sum()
total=part1+part2
new=pd.DataFrame(total).reset_index()
new.rename({'team1': 'Team',0 : 'Total goal'},
          axis = "columns", inplace = True)
```

Create new columns

```python
data =  pd.concat([new,new_1['goal inside the penalty'],new_2['goal outside the penalty area'],new_3['yellow cards'],new_4['red cards']
                   ,new_5['total attempts'],new_6['on target attempts'],new_7['off target attempts'],new_8['attempts inside the penalty area'],new_9['attempts outside the penalty area'],
                   new_10['Conceded'],new_11['Assits'],new_12['Passes completed'],new_13['Corners team'],new_15['Free kicks'],
                   new_16['Completed defensive line breaks'],new_17['Offsides team'],new_18['Total attempted defensive line breaks']],axis=1)
data.columns
```

```python
# Lấy dữ liệu từ dictionary
total_goal = data['Total goal']
goal_inside_penalty = data['goal inside the penalty']
goal_outside_penalty = data['goal outside the penalty area']
on_target_attempts = data['on target attempts']
defensive_line_breaks = data['Completed defensive line breaks']
yellow_cards = data['yellow cards']
red_cards = data['red cards']
off_target_attempts = data['off target attempts']
attempts_inside_penalty = data['attempts inside the penalty area']
attempts_outside_penalty = data['attempts outside the penalty area']
assists = data['Assits']
passes_completed = data['Passes completed']
corners = data['Corners team']
offsides = data['Offsides team']
attempted_defensive_line_breaks = data['Total attempted defensive line breaks']
free_kicks = data['Free kicks']
conceded = data['Conceded']

# Tính điểm số
points = (total_goal * 4) + (goal_inside_penalty * 3) + (goal_outside_penalty * 2) + (on_target_attempts * 0.5) + (defensive_line_breaks * 0.5) + (yellow_cards * -1) + (red_cards * -2) + (off_target_attempts * -0.1) + (attempts_inside_penalty * 0.2) + (attempts_outside_penalty * 0.1) + (assists * 1) + (passes_completed * 0.1) + (corners * 0.2) + (defensive_line_breaks * 0.5) + (offsides * -0.5) + (attempted_defensive_line_breaks * 0.5) + (free_kicks * 0.2) - conceded
```

```python
data = data.assign(points=points)
```

# Train model

```python
import sklearn
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
%matplotlib inline
```

# Process data to train

```python
x = data.drop(['Team','points'], axis= 1)
y = data['points']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=50)
```

# Use LinearRegression, StandardScaler, Polynominal

```python
ListofTuples=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(degree=1)),('model',LinearRegression())]
```

```python
pipe = Pipeline(ListofTuples)
pipe.fit(X_train,y_train)
pipe.score(X_test,y_test)
pipe.predict(X_test)
```

```python
def Predict_Win_Team_LR(x,y):
  score1 = pipe.predict(data[data['Team']==x].drop(['Team','points'],axis =1))
  score2 = pipe.predict(data[data['Team']==y].drop(['Team','points'],axis =1))
  total_score = score1[0] + score2[0]
  percentage1 = int((score1[0]/total_score)*100)
  percentage2 = int((score2[0]/total_score)*100)
  win_percentages = np.array([percentage1, percentage2])
  fig, ax = plt.subplots()
  ax.bar([0, 1], win_percentages)
  ax.set_xticks([0, 1])
  ax.set_xticklabels([x, y])
  ax.set_ylabel('Win percentage')
  plt.show()

```

```python
Predict_Win_Team_LR('SPAIN','WALES')
```

# Use random forest 🌳

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score, mean_squared_error
rf_regressor = RandomForestRegressor(n_estimators=165, random_state=50)
rf_regressor.fit(X_train, y_train)
y_pred_regressor = rf_regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred_regressor)
print(f"Mean squared error: {mse:.2f}")
rf_regressor.score(X_test,y_test)
```

```python
def Predict_Win_Team_RF(x,y):
  score1 = rf_regressor.predict(data[data['Team']==x].drop(['Team','points'],axis =1))
  score2 = rf_regressor.predict(data[data['Team']==y].drop(['Team','points'],axis =1))
  total_score = score1[0] + score2[0]
  percentage1 = int((score1[0]/total_score)*100)
  percentage2 = int((score2[0]/total_score)*100)
  win_percentages = np.array([percentage1, percentage2])
  fig, ax = plt.subplots()
  ax.bar([0, 1], win_percentages)
  ax.set_xticks([0, 1])
  ax.set_xticklabels([x, y])
  ax.set_ylabel('Win percentage')
  plt.show()
```

```python
Predict_Win_Team_RF('ENGLAND','FRANCE')
```

# Use KNN 🤖

```python
from sklearn.neighbors import KNeighborsRegressor
knn_regressor = KNeighborsRegressor(n_neighbors=10)
knn_regressor.fit(X_train, y_train)
y_pred_regressor1 = knn_regressor.predict(X_test)
mse1 = mean_squared_error(y_test, y_pred_regressor1)
print(f"Mean squared error: {mse1:.2f}")
knn_regressor.score(X_test,y_test)
```

```python
def Predict_Win_Team_KNN(x,y):
  score1 = knn_regressor.predict(data[data['Team']==x].drop(['Team','points'],axis =1))
  score2 = knn_regressor.predict(data[data['Team']==y].drop(['Team','points'],axis =1))
  total_score = score1[0] + score2[0]
  percentage1 = int((score1[0]/total_score)*100)
  percentage2 = int((score2[0]/total_score)*100)
  win_percentages = np.array([percentage1, percentage2])
  fig, ax = plt.subplots()
  ax.bar([0, 1], win_percentages)
  ax.set_xticks([0, 1])
  ax.set_xticklabels([x, y])
  ax.set_ylabel('Win percentage')
  plt.show()
```

```python
Predict_Win_Team_KNN('JAPAN','KOREA REPUBLIC')
```

```python
def Predict_Group_State(x,y,z,k):
#use knn
  score1_knn = knn_regressor.predict(data[data['Team']==x].drop(['Team','points'],axis =1))
  score2_knn = knn_regressor.predict(data[data['Team']==y].drop(['Team','points'],axis =1))
  score3_knn = knn_regressor.predict(data[data['Team']==z].drop(['Team','points'],axis =1))
  score4_knn = knn_regressor.predict(data[data['Team']==k].drop(['Team','points'],axis =1))
  total_score_knn = score1_knn[0] + score2_knn[0] + score3_knn[0]+  score4_knn[0]
  percentage1_knn = int((score1_knn[0]/total_score_knn)*100)
  percentage2_knn = int((score2_knn[0]/total_score_knn)*100)
  percentage3_knn = int((score3_knn[0]/total_score_knn)*100)
  percentage4_knn = int((score4_knn[0]/total_score_knn)*100)
# rf
  score1_rf = rf_regressor.predict(data[data['Team']==x].drop(['Team','points'],axis =1))
  score2_rf = rf_regressor.predict(data[data['Team']==y].drop(['Team','points'],axis =1))
  score3_rf = rf_regressor.predict(data[data['Team']==z].drop(['Team','points'],axis =1))
  score4_rf = rf_regressor.predict(data[data['Team']==k].drop(['Team','points'],axis =1))
  total_score_rf = score1_rf[0] + score2_rf[0] + score3_rf[0]+  score4_rf[0]
  percentage1_rf = int((score1_rf[0]/total_score_rf)*100)
  percentage2_rf = int((score2_rf[0]/total_score_rf)*100)
  percentage3_rf = int((score3_rf[0]/total_score_rf)*100)
  percentage4_rf = int((score4_rf[0]/total_score_rf)*100)
# lr
  score1_lr = pipe.predict(data[data['Team']==x].drop(['Team','points'],axis =1))
  score2_lr = pipe.predict(data[data['Team']==y].drop(['Team','points'],axis =1))
  score3_lr = pipe.predict(data[data['Team']==z].drop(['Team','points'],axis =1))
  score4_lr  = pipe.predict(data[data['Team']==k].drop(['Team','points'],axis =1))
  total_score_lr  = score1_lr [0] + score2_lr[0] + score3_lr[0]+  score4_lr[0]
  percentage1_lr = int((score1_lr[0]/total_score_lr)*100)
  percentage2_lr = int((score2_lr[0]/total_score_lr)*100)
  percentage3_lr = int((score3_lr[0]/total_score_lr)*100)
  percentage4_lr = int((score4_lr[0]/total_score_lr)*100)

  #draw pie chart
  fig, axs = plt.subplots(1,3,figsize=(30, 30))

  variables = [x, y, z, k]
  values1 = [percentage1_knn, percentage2_knn, percentage3_knn, percentage4_knn]
  values2 = [percentage1_rf, percentage2_rf, percentage3_rf, percentage4_rf]
  values3 = [percentage1_lr, percentage2_lr, percentage3_lr, percentage4_lr]
  colors = ['blue', 'green', 'yellow', 'red']
  explode = (0, 0, 0, 0)  # Explode the third slice


  axs[0].pie(values1, labels=variables, radius=5, textprops={'fontsize': 30})
  axs[0].set_title('Pie Chart KNN', fontsize=50, y=3.5)
  axs[1].pie(values2, labels=variables, radius=5, textprops={'fontsize': 30})
  axs[1].set_title('Pie Chart RF', fontsize=50, y=3.5)
  axs[2].pie(values3, labels=variables, radius=5, textprops={'fontsize': 30})
  axs[2].set_title('Pie Chart LR', fontsize=50, y=3.5)

  # Adjust the space between the subplots
  fig.subplots_adjust(wspace=5)

  # Add a title to the figure
  fig.suptitle('Win percentage', fontsize=100, y=0.8, fontweight='bold')

  # Display the plot
  plt.show()
```
